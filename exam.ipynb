```python
import numpy as np
import math
import matplotlib as mpl
from matplotlib import pyplot as plt

#Question 1

def question1(N):
    n = 2
    a = 1 #a0
    A = 1 #a1
    suite = a + A
    while suite <= N:
        n += 1
        a, A = A, suite
        suite = a + A
    return n + 1

print("Le premier n naturel tels que An >= 10000 est :",question1(10000))


def quotients(n):
    fig = plt.figure()
    axis = fig.add_subplot(1, 1, 1)

    a0, a1 = 1, 1
    Qk = [1]

    for i in range(n):
        temp = a0 + a1
        Qk.append(temp / a1)
        a0, a1 = a1, temp

    axis.plot(range(n+1), Qk)
    axis.set_title('Qk = Ak / Ak-1')
    plt.show()
    return Qk

print(quotients(1000))

def suite(n):
  A = [1, 1]
  a0, a1 = 1, 1

  for i in range(n):
    temp = a0 + a1
    A.append(temp)
    a0, a1 = a1, temp

  return A


def limite(n):
  fig = plt.figure()
  axis = fig.add_subplot(1, 1, 1)
  A = suite(n)
  Q = 1.61
  Qk = []
  for i in range(n+2):
    temp = Q ** i
    Qk.append(temp)
  axis.plot(range(n+2), A, color='k')
  axis.plot(range(n+2), Qk, color='r')
  axis.set_title(f'Conjecture de Q qui serait égale à {Q}')

print(limite(100))

def equivalent(n):
  fig = plt.figure()
  axis = fig.add_subplot(1, 1, 1)
  A = suite(n)
  Q = 1.618033988749895     #Q = 1 + an-2 / an-1
  Qk = []
  for i in range(n+2):
    temp = Q ** i
    Qk.append(temp)
  axis.plot(range(n+2), A, color='k')
  axis.plot(range(n+2), Qk, color='r')
  axis.set_title(f'Equivalent de An en + inf avec Q = {Q}')

print(equivalent(1000))

```

```python
import numpy as np
import math
import matplotlib as mpl
from matplotlib import pyplot as plt


def terme_suivant(x0, y0):
  xn = 1 - 1.4 * (x0**2) + 0.3 * y0
  yn = x0
  return xn, yn

def suite_H(x0, y0, N):
  P = [[x0, y0]]
  for i in range(N):
    x0, y0 = terme_suivant(x0, y0)
    P.append([x0, y0])
  return P

print(suite_H(0, 1, 2))

#Je n'ai pas utilisé ma fonction suite_H car j'avais sauté la question donc j'ai tout refait moi même sur esquisse

def esquisse_suite_H(x0, y0, N):
  fig = plt.figure()
  axis = fig.add_subplot(1, 1, 1)
  X, Y = [x0], [y0]
  for i in range(N):
    x0, y0 = terme_suivant(x0, y0)
    X.append(x0)
    Y.append(y0)

  axis.plot(X, Y, color='k', marker='x', linestyle='none')
  axis.grid()
  axis.set_title(f'Esquisse de la suite pour n égale à {N}')
  axis.set_xlabel('Abscisse du point P')
  axis.set_ylabel('Ordonnée du point P')
  return None

print(esquisse_suite_H(1.1, -0.2, 1000))
print(esquisse_suite_H(0.1, 0.1, 1000))

def terme_proche_H(x0, y0, xM, yM, eps):
  n = 0
  N = []
  xvec = (xM - x0)**2
  yvec = (yM - y0)**2
  PkM = (xvec + yvec)**0.5
  for i in range(2):  #Pour déterminer le Second terme tel que la distance de Pk à M > epsilon
    while PkM >= eps :
      x0, y0 = terme_suivant(x0, y0)
      xvec = (xM - x0)**2
      yvec = (yM - y0)**2
      PkM = (xvec + yvec)**0.5
      n += 1
    PkM = eps + 1 #Réagrandit PkM pour rentrer dans la boucle while
    N.append([n, [x0, y0]])
  return N


print(terme_proche_H(0.1,0.1, 0.12, 0.8, 0.01))

```
