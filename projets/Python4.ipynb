```python
import numpy as np
import math
import matplotlib as mpl
from matplotlib import pyplot as plt

#np.shape = la dimension de la matrice
#np.array = matrice de n lignes
#np.size = le nombre d'élément
#np.shape(M.T) = T renvoie la transposée de la matrice M
#M.dot(A) = produit de la matrice A et M
#M[y,x] = affiche le nombre à la ligne x et colonne y
#M[x,:] = renvoie la x-ieme ligne
#M[x:,y] = renvoie les nombres de la y-ieme colonne après la x-ieme ligne
#np.zeros(n) = renvoie une matrice 1 dimension de n zéros
#np.eye(n) = renvoie une matrice identitée
#np.linspace(x, y, z) = renvoie une matrice commençant à x allant jusqu'à y en ayant z valeurs au total en additionnant le même nombre


a = 30
b = 15
def euclide_bezout(a,b):
  E = []
  while b != 0 :
    L = [a, b, a//b, a%b]
    E.append(L)
    a = L[1]
    b = L[3]
  J = np.array(E)
  x,y = np.shape(J)
  d = int(J[x-2, y-1])
  return d


def Bezout(a,b):
  d = euclide_bezout(a,b)
  v = 1
  while ((v*b-d)%a != 0):
    v += 1
  u = int((d-v*b)/a)
  return u, v, d

print(Bezout(a,b))
```

```python
import numpy as np
import matplotlib.pyplot as plt


c = 3.2

def f_log(x):
    return c * x * (1 - x)

def f2(x):
    return f_log(f_log(x))


def suite(a, N):
    u = [a]
    for _ in range(N):
        u.append(f_log(u[-1]))
    return u


def sous_suites(U):
    v = U[::2]
    w = U[1::2]
    return v, w

def rpz_escalier(m, M, a, N):
    fig = plt.figure()
    axis = fig.add_subplot(1, 1, 1)
    axis.grid()
    X = np.linspace(m, M, 500)
    Y = f_log(X)
    axis.plot(X, Y, label="f(x)")
    axis.plot(X, X, linestyle="dashed", color="red", label="y = x")

    u = a
    for j in range(N):
        axis.plot([u, u], [u, f_log(u)], color="black")
        axis.plot([u, f_log(u)], [f_log(u), f_log(u)], color="black")
        u = f_log(u)
    plt.legend()
    plt.show()
    plt.close(fig)

def trace_fonc(c, a, N):
    global f_log
    f_log = lambda x: c * x * (1 - x)

    fig = plt.figure()
    axis = fig.add_subplot(1, 1, 1)
    axis.grid()
    X = np.linspace(0, 1, 500)
    Y2 = f2(X)

    U = suite(a, 2*N)
    V, W = sous_suites(U)

    axis.plot(X, Y2, linestyle="dashed")
    axis.plot(X, X, color="gray", linestyle="dotted")

    axis.plot(V, V, 'o')
    axis.plot(W, W, 'x')

    plt.legend()
    plt.show()
    plt.close(fig)



m, M, a, N = 0, 1, 0.4, 50
rpz_escalier(m, M, a, N)
trace_fonc(c, a, N)
```

```python
from numpy import linspace
from matplotlib.pyplot import scatter, show, xlim, ylim

def f(r, x):
    return r * x * (1 - x)

def feigenbaum(r1, r2, nb, N):
    r_range = linspace(r1, r2, nb)
    x = []
    y = []
    for r in r_range:
        u = .1
        for _ in range(N):
            u = f(r, u)
        for _ in range(N):
            u = f(r, u)
            x.append(r)
            y.append(u)
    scatter(x, y, marker='.', s=1)
    xlim(r1, r2)
    ylim(0, 1)
    show()

print(feigenbaum(0.1, 4, 2000, 2000))
```

```python
import numpy as np
import math
import matplotlib as mpl
from matplotlib import pyplot as plt

def simulation_TB(a, b, c, d, T, B, N):
  fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

  mT = [T]
  mB = [B]
  for i in range(N):
    tmp = (1 + a - a * T / 3000) * T - b * T * B
    B = B * T * c * b + (1 - d) * B
    T = tmp
    mB.append(B)
    mT.append(T)
# Tracé en fonction du temps
  ax1.plot(range(N+1), mT, label='Truites')
  ax1.plot(range(N+1), mB, label='Brochets')
  ax1.set_title("Évolution des populations")
  ax1.set_xlabel("Temps")
  ax1.set_ylabel("Population")
  ax1.legend()

# Tracé B(T)
  ax2.plot(mT, mB, 'o-', color='green')
  ax2.set_title("Brochets en fonction des truites")
  ax2.set_xlabel("Truites")
  ax2.set_ylabel("Brochets")

  plt.tight_layout()
  plt.show()

  return mT, mB

print(simulation_TB(0.4, 0.021, 0.015, 0.25, 30, 52, 10))
```

```python
import numpy as np
import math
import matplotlib as mpl
from matplotlib import pyplot as plt
import numpy.random as npr

def f(x):
  return x * (1-x) * np.sin(200 * x * (1-x))**2

def intMontCarlo( xmax, xmin, n):
  fig = plt.figure()
  axis = fig.add_subplot(1, 1, 1)
  X = np.linspace(xmin, xmax, n)
  Y = f(X)
  Y0 = [ f(npr.rand() * (xmax-xmin) + xmin) for k in range(n)]
  S = sum(Y0)
  I = (xmax-xmin)/n*S
  if I < 0 :
    I = - I
  axis.plot(X,Y)
  return I

def Simpson( a, b, n):
  h = (b - a)/n
  Y = []
  for k in range(n):
    Y.append(((h/6) * (f(a + k * h) + 4 * f(a + (k + 1/2) * h) + f(a + (k+1) * h ))))
  I = sum(Y)
  return I



print(intMontCarlo(0 , 1 , 10000))
print(Simpson(0, 1, 70))
```

```python
import numpy as np
import math
import matplotlib as mpl
from matplotlib import pyplot as plt
import numpy.random as npr

def dessin_parquet(R, L, l):
  fig = plt.figure()
  axis = fig.add_subplot(1, 1, 1, aspect ='equal')
  X = np.array([-L, L])/2
  for i in range(R):
    Y = (i - R//2) * l * np.array([1, 1])
    axis.plot(X, Y, c='k', lw=.7)

  axis.set_xlim(-(L/2 + l), (L/2 + l))
  axis.set_ylim(-R//2 * l , R//2 * l + l/2)

  return axis, fig



def aiguille(xC, yC, theta, d):
  '''
  créé une aiguille de taille centre xC yC et fait une rotation theta de longueur d
  sur le parquet de dessin_parquet

  '''
  delta = d / 50
  x = np.array([0, 0, d, 0]) - d / 2
  y = np.array([-delta /2, delta/2, 0, -delta/2])
  X = x * np.cos(theta) - y * np.sin(theta) + xC
  Y = x * np.sin(theta) + y * np.cos(theta) + yC

  return X, Y


def experiment_graph(nbL, d, R, L, l):
  '''
  Lance nbL aiguille sur le parquet
  on ajoute un au compteur si le reste de la division euclidienne (rabatte entre
  deux rainures sur l'axe des abscisses) est inférieur
  ou égale à la moitié de la taille de l'aiguille fois sin(theta)
  '''
  axis, fig = dessin_parquet(R, L, l)
  compteur = 0
  for i in range(nbL):
    xC = npr.uniform(-L/2, L/2)
    yC = npr.uniform(-R//2 * l , R//2 * l)
    theta = npr.uniform(0, np.pi)
    X, Y = aiguille(xC, yC, theta, d)
    r = abs(yC % l)
    if r <= abs((d / 2) * np.sin(theta)):
      compteur += 1
      axis.plot(X, Y, c='r')
    else :
      axis.plot(X , Y, c='b')
  return compteur



def PideN(n, a, R, L, l):
    F = experiment_graph(n, a, R, L, l)
    if F == 0:
        return None
    return (a * n) / (l * F)

pi_estime = PideN(2000, 1.5, 9, 10, 1.7)
print("Estimation de π :", pi_estime)
```

```python
import matplotlib.pyplot as plt
import math

def koch(p1, p2, d):
    if d == 0:
        return [p1, p2]
    else:
        x1, y1 = p1
        x5, y5 = p2

        dx = (x5 - x1) / 3
        dy = (y5 - y1) / 3

        x2, y2 = x1 + dx, y1 + dy
        x3, y3 = 0.5 * (x1 + x5) - math.sqrt(3) * (y1 - y5) / 6, 0.5 * (y1 + y5) - math.sqrt(3) * (x5 - x1) / 6
        x4, y4 = x1 + 2 * dx, y1 + 2 * dy

        return    (koch((x1, y1), (x2, y2), d - 1)[:-1] + #retourne la récursive et supprime le doublon de profondeur d - 1
                  koch((x2, y2), (x3, y3), d - 1)[:-1] +
                  koch((x3, y3), (x4, y4), d - 1)[:-1] +
                  koch((x4, y4), (x5, y5), d - 1))
def dessin(d):
    p1 = (0, 0)
    p2 = (1, 0)

    points = koch(p1, p2, d)

    x, y = zip(*points) #Converti la liste de points en multiples coordonées


    plt.figure(figsize=(8, 4))
    plt.plot(x, y, color='blue')
    plt.axis('equal')
    plt.axis('off')
    plt.title(f"Flocon de Koch - profondeur {d}")
    plt.show()


dessin(5)
```

```python
import random

def choisir_mot():
    mots = ["pyhton"]
    return random.choice(mots)

def afficher_mot(mot, lettres_trouvees):
    return " ".join([lettre if lettre in lettres_trouvees else "_" for lettre in mot])

def pendu():
    mot = choisir_mot()
    lettres_trouvees = set()
    lettres_ratees = set()
    essais_restants = 8

    print("Bienvenue dans le jeu du Pendu !")

    while essais_restants > 0:
        print("\nMot à deviner :", afficher_mot(mot, lettres_trouvees))
        print(f"Lettres ratées : {', '.join(sorted(lettres_ratees))}")
        print(f"Essais restants : {essais_restants}")

        lettre = input("Proposez une lettre : ").lower()

        if not lettre.isalpha() or len(lettre) != 1:
            print("Veuillez entrer une seule lettre valide.")
            continue

        if lettre in lettres_trouvees or lettre in lettres_ratees:
            print("Vous avez déjà proposé cette lettre.")
            continue

        if lettre in mot:
            lettres_trouvees.add(lettre)
            print("Bonne lettre !")
        else:
            lettres_ratees.add(lettre)
            essais_restants -= 1
            print("Raté !")

        if all(l in lettres_trouvees for l in mot):
            print("\nFélicitations ! Vous avez trouvé le mot :", mot)
            break
    else:
        print("\nDommage, vous avez perdu. Le mot était :", mot)

if __name__ == "__main__":
    pendu()
```

```python
import numpy as np
import math
import matplotlib.pyplot as plt
import numpy.random as npr

def module(z):
    return abs(z)

def fonction(z, c):
    return z**2 + c

def suite(a, c, N):
    fig = plt.figure(figsize=(6, 2))
    axis = fig.add_subplot(1, 1, 1, aspect='auto')
    u = []
    z = a
    for i in range(N):
        z = fonction(z, c)
        mod = module(z)
        u.append(mod)

    axis.plot(range(N), u, marker='x', color='k')
    return u

def max_iter(Nmax, s, c):
  z = c
  for n in range(Nmax):
    if module(z) > s :
      return n
    z = z ** 2 + c
  return Nmax

def max_iter(Nmax, s, a, c):
  z = a
  N = 0
  while N < Nmax :
    if abs(z) > s :
      return N

    z = fonction(z, c)
    N += 1
  return Nmax


def julia(Nmax, s, Ndiv, r, c):
    h = 2 * r / (Ndiv - 1)
    M = np.zeros([Ndiv, Ndiv])

    for j in range(Ndiv):
        for k in range(Ndiv):
            w = complex(-r + k * h, -r + j * h)
            M[j, k] = min(1/255 * max_iter(Nmax, s, w, c), 1)

    plt.imshow(M, cmap='hot', origin='lower', extent=[-r, r, -r, r])
    plt.colorbar()
    plt.title("Ensemble de Julia")
    plt.show()
    return M

julia(1000, 100, 150, 1.5, -0.79 + 0.156j)
```
